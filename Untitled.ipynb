```python
html_list = []
for num in tqdm(range(17300000, 17311831)):
    if len(html_list) == 5000:
        with open(f"D:\\clien\\{num}.pkl", "wb") as f:
            pkl.dump(html_list, f) 
        del html_list
        html_list = []
    target_url = f"https://www.clien.net/service/board/park/{num}"
    try:
        searchpage_html = requests.get(target_url, headers={'User-Agent': ua.chrome})
    except:
        continue
    aa = BeautifulSoup(searchpage_html.text, "html.parser")
    if "해당 페이지가 삭제되었거나 잘못된 접근일 수 있습니다." in aa.text:
        time.sleep(0.1)
        continue
    if len(aa.select('div[class="content_view"]')) == 0:
        time.sleep(0.1)
        continue
    html_list.append((target_url, str(aa.select('div[class="content_view"]')[0])))
    time.sleep(0.05)
    
with open(f"D:\\clien\\{num}.pkl", "wb") as f:
    pkl.dump(html_list, f)
del html_list
```
